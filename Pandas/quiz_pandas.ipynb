* 5) Remplacer les valeurs manqquantes dans la colonne Age apr la médiane de cette colonne

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

* 6) Remplir les valeurs manquantes dans Salaire en utilisant la moyenne par département 

```python
df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x : x.fillna(x.mean()))

df.head(20)
```

* 7) Convertir toutes les colonnes numériques en types approprié (float ou int)

```python
df['Age'] = np.int_(df['Age'])
df['Salary'] = np.float32(df['Salary'])
```

* 8) Rmeplacer les valeurs 'YES/NO' dans Remote par 'Oui/Non'

```python
df['Remote'] = df['Remote'].map({'Yes': 'Oui', 'No': 'Non'})

df['Remote'].head(5)
```

* 9) Créer une nouvelle colonne Ancienneté_Categorie qui classe les années d'expérience en :
  * Junior : < 3 ans
  * Intermédiaire : 3-7 ans
  * Senior : 8-15 ans
  * Expert : > 15 ans

```python
def check(x):
    if x < 3 : 
        return 'Junior'
    elif 3 <= x <= 7 :
        return 'Intermédiaire'
    elif 8 <= x <= 15 :
        return 'Senior'
    else:
        return 'Expert'
    
    
df['Anciennete_Categorie'] = df['Years_Experience'].apply(lambda x : check(x))

df.head(5)
```

* 10) Calculer le salaire moyen global :

```python
moy_global_salary = df['Salary'].mean()

print(f"Le salaire moyen global : {moy_global_salary}  ")
```

* 11) Trouver l'amployé(e) avec le salaire le plus élevé

```python
max_salary = df.loc[df['Salary'].idxmax()]
print(f"l'amployé(e) avec le salaire le plus élevé :\n {max_salary}")
```

* 12) Calculer le salaire moyen par département .
  

```python
salary_moy_departement = df.groupby('Department')['Salary'].mean()

salary_moy_departement
```

* 13) Claculer la moyenne et la médian des salaires par groupe d'ancienneté
  

```python
mean_median_salary = df.groupby('Anciennete_Categorie').agg({
    'Salary' : ['mean', 'median']
})

mean_median_salary
```

* 14) Compter combien d'amployés travaillent en télétravail (Remote) par département.

```python
df_14 = df[['Department', 'Remote']].copy()

df_14['Remote'] = df_14['Remote'].map({'Oui': 1, 'Non': 0})

remote_count = df_14.groupby('Department')['Remote'].sum()

print(remote_count)
```
