# Gestion des permissions : 

 
 * chmod : Modifier les droits d’acc`es d’un fichier ou dossier.
 
 * chown : Changer le propri´etaire et/ou le groupe d’un fichier ou dossier.
 
 * sudo : Permet d’ex´ecuter une commande en tant qu’administrateur
 
 * su: Change d’utilisateur (switch user ) dans le terminal.
 
 * passwd : Permet de changer le mot de passe d’un utilisateur.
 
 * adduser et deluser : G`ere les comptes utilisateurs .
 
 * groups : Affiche les groupes auxquels appartient un utilisateur.
 
 * addgroup / delgroup : Cr´eer ou supprimer un groupe .
 
 * whoami : Affiche le nom de l’utilisateur actuel connect´e au terminal.


1. Les Bases des Permissions
Structure : [type][propriétaire][groupe][autres]

Exemple : -rwxr-xr--

'-' : Type de fichier (- = fichier, d = dossier)

'rwx' : Propriétaire (read, write, execute) `7`

'r-x' : Groupe (read, execute) `5`

'r--' : Autres (read seulement) `4`







#### 1. chmod : 
Modifier les permissions
Syntaxe : `chmod [qui][opérateur][permission] fichier`




* Méthode Symbolique :

Ajouter permission exécution au propriétaire : 
`chmod u+x script.sh`

Retirer écriture au groupe et autres : 
`chmod go-w fichier.txt`

Donner toutes permissions au propriétaire : 
`chmod u=rwx document.pdf`

* Méthode Numérique (octale) :

755 : rwxr-xr-x : `chmod 755 script.sh`

644 : rw-r--r-- : `chmod 644 config.txt`

700 : rwx------ (seul le propriétaire) : `chmod 700 secret.txt`

#### 2. chown 
- Changer propriétaire/groupe
Syntaxe : `chown utilisateur:groupe fichier`

* Changer propriétaire : 

`chown alice fichier.txt`

* Changer propriétaire et groupe : 

`chown alice:developers script.py`

* Changer seulement le groupe : 

`chown :admins backup.log`

* Récursif pour un dossier : 

`chown -R bob:team /projet`/

#### 3. sudo 
- Exécuter en administrateur

* Mettre à jour le système : 

`sudo apt update`

* Éditer un fichier système : 

`sudo nano /etc/hosts`

* Redémarrer un service : 

`sudo systemctl restart apache2`

* installer un package : 

`sudo apt install nom_pachage`

#### 4. su
- Change d’utilisateur (switch user ) dans le terminal.

* Devenir root : 
`su -`

* Devenir utilisateur "alice" : 
`su alice`

* Retourner à l'utilisateur précédent : 
`exit`

#### 5. passwd :
- Changer mot de passe

* Changer son propre mot de passe

`passwd`

* Changer le mot de passe d'un autre utilisateur

`sudo passwd alice`

#### 6. adduser / deluser : 

* Créer un nouvel utilisateur : 

`sudo adduser bob`

* Supprimer un utilisateur : 

`sudo deluser bob`

* Supprimer avec son home directory : 

`sudo deluser --remove-home bob`

#### 7. groups - Voir les groupes

* Groupes de l'utilisateur courant : 
`groups`

* Groupes d'un utilisateur spécifique : 
`groups alice`

#### 8. addgroup / delgroup

* Créer un nouveau groupe : 
`sudo addgroup developers`

* Supprimer un groupe : 
`sudo delgroup developers`

* Ajouter un utilisateur à un groupe : 
`sudo adduser alice developers`

#### 9. whoami - Utilisateur actuel

* Savoir qui je suis : 
`whoami`

* Combien avec id (plus d'infos) : 
`id`
