## Challenge 2: Nettoyage des Données

#### 1. Compter les valeurs manquantes pour chaque colonne


```python
manquantes = df.isnull().sum()
manquantes
```

#### 2. Remplacer les valeurs manquantes dans 'age' par la médian


```python
df['age'].isnull().sum()
```

```python
df['age'] = df['age'].fillna(
    df['age'].median()
)
```

```python
df['age'].isnull().sum()
```

#### 3. Supprimer la colonne 'deck' (trop de valeurs manquantes)


```python
df.columns
```

```python
df = df.drop('deck', axis=1)
df.columns
```

#### 4. Supprimez les doublons potentiels.


```python
print(f"Avant : ", len(df))
df = df.drop_duplicates()
print(f"Apres : ", len(df))
```

#### 5. Créer une nouvelle colonne 'age_group': 'enfant', 'adulte', 'senior'

```python
def check(x):
    if x < 18 : 
        y = 'enfant'
    elif 18 <= x < 40:
        y = 'adulte'
    else:
        y = 'senior'
    return y

df['age_group'] = df['age'].apply(lambda x : check(x))

df.head(10)

```

```python
print(df['age_group'].value_counts())
```

## Challenge 3: Agrégations et GroupBy

#### 1. Calculer le taux de survie global


```python
taux_survie_global = df['survived'].mean() * 100
print(f"taux_survie_global : {taux_survie_global:.2f}%")
```

```python
survival_counts = df['survived'].value_counts(normalize=True)
print(f"\nRépartition survie (value_counts):")
print(survival_counts * 100)
```

#### 2. Taux de survie par classe


```python
taux_survie_par_class = df.groupby('pclass')['survived'].mean() * 100
taux_survie_par_class
```

```python
survie_par_classe = df.groupby('pclass').agg({
    'survived': ['mean', 'sum', 'count']
})
print(survie_par_classe)
```

#### 3. Taux de survie par sexe


```python
taux_survie_par_sex = df.groupby('sex')['survived'].mean() * 100
taux_survie_par_sex
```

#### 4. Âge moyen par classe et sexe


```python
age_moy_par_class_sex = df.groupby(['pclass', 'sex'])['age'].mean()
age_moy_par_class_sex
```

```python
for (pclass, sex), moy in age_moy_par_class_sex.items():
    print(f"({pclass}, {sex}) = {moy:.2f}")
```

#### 5. Nombre de survivants par port d'embarquement

```python
nbr_survived_par_embarked = df.groupby('embarked')['survived'].sum()

nbr_survived_par_embarked
```

```python
nbr_sur_par_embarked = df.groupby('embarked').agg({
    'survived': ['mean', 'sum', 'median', 'count'],
    'fare': ['mean', 'sum', 'median', 'count'],
    'age': ['mean', 'sum', 'median', 'count'],
})

nbr_sur_par_embarked
```

## Challenge 4: Manipulation de Données

#### 1. Trier le DataFrame par âge décroissant


```python
df_sorted = df.sort_values('age', ascending=True)

df_sorted.head(15)
```

#### 2. Ajouter une colonne 'family_size' = sibsp + parch + 1 avec apply()


```python
df['family_size'] = df.apply(lambda x : x['sibsp'] + x['parch'] + 1 , axis=1)

df
```

#### 3. Créer une colonne 'is_alone' (family_size == 1)


```python
df['is_alone'] = df['family_size'].apply(lambda x : True if x == 1 else False)

df.head(10)
```

#### 4. Utilisez map() pour mapper des valeurs catégorielles (ex. : mapper sex de 'male'/'female' à 'M'/'F').


```python
df['sex'] = df['sex'].map({'male': 'M', 'female': 'F' })

df.head(6)
```

#### 5. Augmenter de 10% le prix des billets des survivants de 1ère classe (avec loc[])


```python
mask = ((df['survived'] == 1) & (df['pclass'] == 1))


print(f"Avant : \n{df.loc[mask, 'fare'].head(5)}\n")
df.loc[mask, 'fare'] = df.loc[mask, 'fare'] * 1.10
print(f"Apres : \n{df.loc[mask, 'fare'].head(5)}\n")


```

#### 6. Marquer comme 'VIP' les passagers avec billet > 200 (avec loc[])

```python
df['vip_status'] = 'Standard'
df.loc[df['fare'] > 200, 'vip_status'] = 'VIP'

print(df['vip_status'].value_counts())

```

#### 1. Histogramme des âges

```python
# Solution
import matplotlib.pyplot as plt

# Histogramme simple
plt.figure(figsize=(10, 6))
df['age'].hist(bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribution des âges des passagers du Titanic')
plt.xlabel('Âge')
plt.ylabel('Nombre de passagers')
plt.grid(axis='y', alpha=0.3)
plt.show()

# Explication:
# - hist() crée un histogramme directement depuis une série pandas
# - bins=30: nombre de barres (intervalles d'âge)
# - alpha: transparence pour mieux voir les superpositions
# - edgecolor: couleur des bordures des barres

# Version améliorée avec analyse par groupe
plt.figure(figsize=(12, 6))

# Histogramme comparatif survivants vs non-survivants
df[df['survived'] == 1]['age'].hist(bins=30, alpha=0.7, label='Survivants', color='green')
df[df['survived'] == 0]['age'].hist(bins=30, alpha=0.5, label='Décédés', color='red')

plt.title('Distribution des âges: Survivants vs Décédés')
plt.xlabel('Âge')
plt.ylabel('Nombre de passagers')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Statistiques descriptives
print("Statistiques d'âge:")
print(f"Âge moyen: {df['age'].mean():.1f} ans")
print(f"Âge médian: {df['age'].median():.1f} ans")
print(f"Âge minimum: {df['age'].min():.1f} ans")
print(f"Âge maximum: {df['age'].max():.1f} ans")
```

#### 2. Diagramme en barres du nombre de survivants par classe


```python
plt.figure(figsize=(10, 6))

survie_par_classe = df.groupby('pclass')['survived'].value_counts().unstack()
survie_par_classe = survie_par_classe.rename(columns={0:"mato", 1:"hyin"})

survie_par_classe.plot(kind='bar', color=['red', 'green'], alpha=0.3)
plt.title('Nombre de survivants par classe')
plt.xlabel('Classe')
plt.ylabel('Nombre de passagers')
plt.xticks(rotation=0)
plt.legend(title='Statut')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

#### 3. Heatmap de corrélation entre les variables numériques

```python
plt.figure(figsize=(10, 8))

numerical_cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
# numerical_cols = df.columns
correlation_matrix = df[numerical_cols].corr()

# Heatmap
import seaborn as sns
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='coolwarm', 
            center=0,
            square=True,
            fmt='.2f',
            cbar_kws={'shrink': 0.8})

plt.title('Heatmap de corrélation entre variables numériques')
plt.tight_layout()
plt.show()

# Explication:
# - corr() calcule la matrice de corrélation de Pearson
# - annot=True: affiche les valeurs dans les cases
# - cmap='coolwarm': échelle de couleurs bleu(negatif) -> rouge(positif)
# - center=0: le zéro est au centre de l'échelle de couleurs
```

#### 4. Boxplot de la distribution des prix par classe

```python
plt.figure(figsize=(12, 6))

df.boxplot(column='fare', by='pclass', grid=False)
plt.title('Distribution du prix des billets par classe')
plt.suptitle('')
plt.xlabel('Classe')
plt.ylabel('Prix du billet ($)')
plt.show()

# Explication:
# - boxplot() montre la distribution avec médiane, quartiles et outliers
# - by='pclass': groupe par classe
# - On voit bien la différence de prix entre les classes
```
